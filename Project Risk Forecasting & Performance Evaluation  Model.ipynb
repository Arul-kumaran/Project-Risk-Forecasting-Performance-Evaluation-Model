### STEP 2: DATA CLEANING & PREPROCESSING

```python
# Create a copy of the dataset for cleaning
df_clean = df.copy()
```

### 1. Remove duplicates

```python
print("\nDuplicate rows before:", df_clean.duplicated().sum())
df_clean.drop_duplicates(inplace=True)
print("Shape after removing duplicates:", df_clean.shape)
```

### 2. Handle missing values

```python
df.isnull().sum()
```

```python
print("\nMissing values before:")
print(df_clean.isnull().sum()[df_clean.isnull().sum() > 0])
```

```python
df_clean['Tech_Environment_Stability'] = df_clean['Tech_Environment_Stability'].fillna('Unknown')
df_clean['Change_Control_Maturity'] = df_clean['Change_Control_Maturity'].fillna('Basic')
df_clean['Risk_Management_Maturity'] = df_clean['Risk_Management_Maturity'].fillna('Basic')
```

```python
# Numeric columns â€“ convert and handle if any become NaN
num_cols = ['Complexity_Score', 'Project_Budget_USD', 'Team_Size']

for col in num_cols:
    df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')
```

```python
# Final missing value check
print("\nMissing values after cleaning:")
print(df_clean.isnull().sum()[df_clean.isnull().sum() > 0])
```

### 3. Correct data types

```python
df_clean['Complexity_Score'] = pd.to_numeric(df_clean['Complexity_Score'], errors='coerce')
df_clean['Project_Budget_USD'] = pd.to_numeric(df_clean['Project_Budget_USD'], errors='coerce')
df_clean['Team_Size'] = pd.to_numeric(df_clean['Team_Size'], errors='coerce').astype('Int64')
```

```python
# convert some columns to 'category' for efficient analysis
cat_cols = [
    'Tech_Environment_Stability',
    'Change_Control_Maturity',
    'Risk_Management_Maturity',
    'Risk_Level',
    'Project_Type'
]
for col in cat_cols:
    df_clean[col] = df_clean[col].astype('category')
```

```python
print("\nData types after correction:")
print(df_clean.dtypes)
```

### 4. Create derived column

```python
# Budget per team member
df_clean['Budget_per_Team_Member'] = df_clean['Project_Budget_USD'] / df_clean['Team_Size']
```

```python
# Budget category (Low / Medium / High)
df_clean['Budget_Category'] = pd.cut(
    df_clean['Project_Budget_USD'],
    bins=[0,
          df_clean['Project_Budget_USD'].quantile(0.33),
          df_clean['Project_Budget_USD'].quantile(0.66),
          df_clean['Project_Budget_USD'].max()],
    labels=['Low', 'Medium', 'High'],
    include_lowest=True
)
```

```python
# Complexity Category from Complexity_Score
df_clean['Complexity_Category'] = pd.cut(
    df_clean['Complexity_Score'],
    bins=[0,
          df_clean['Complexity_Score'].quantile(0.33),
          df_clean['Complexity_Score'].quantile(0.66),
          df_clean['Complexity_Score'].max()],
    labels=['Low', 'Medium', 'High'],
    include_lowest=True
)
```

```python
# High Risk Flag (1 = High risk, 0 = others)
df_clean['High_Risk_Flag'] = df_clean['Risk_Level'].apply(
    lambda x: 1 if x == 'High' or x == 'High ' else 0
)
```

```python
print("\nSample of derived columns:")
print(df_clean[['Project_Budget_USD', 'Team_Size', 'Budget_per_Team_Member',
                'Complexity_Score', 'Complexity_Category',
                'Risk_Level', 'High_Risk_Flag']].head())
```

### 5.Standardizing Text Columns

```python
text_cols = [
    'Tech_Environment_Stability',
    'Change_Control_Maturity',
    'Risk_Management_Maturity',
    'Risk_Level',
    'Project_Type'
]

for col in text_cols:
    df_clean[col] = (
        df_clean[col]
        .astype(str)
        .str.strip()                       # remove leading/trailing spaces
        .str.replace(r'\s+', ' ', regex=True)  # remove extra spaces
        .str.title()                       # convert to Proper Case
    )
```

```python
print("\nUnique values after cleaning (Tech_Environment_Stability):")
print(df_clean['Tech_Environment_Stability'].unique())
```

```python
print("\nUnique values after cleaning (Risk_Level):")
print(df_clean['Risk_Level'].unique())
```

### 6.Filtering example

```python
# High risk projects
high_risk_projects = df_clean[df_clean['Risk_Level'] == 'High']

print("\nTotal High Risk Projects:", high_risk_projects.shape[0])
```

```python
# High budget & high complexity projects
high_budget_high_complex = df_clean[
    (df_clean['Budget_Category'] == 'High') &
    (df_clean['Complexity_Category'] == 'High')
]
print("High Budget & High Complexity Projects:", high_budget_high_complex.shape[0])
```

```python
# Projects in Unstable tech environment
unstable_env_projects = df_clean[df_clean['Tech_Environment_Stability'] == 'Unstable']
print("Projects in Unstable Tech Environment:", unstable_env_projects.shape[0])
```

### 7.Aggregation Examples

```python
print("\nProject count by Risk Level:")
print(df_clean['Risk_Level'].value_counts())
```

```python
print("\nAverage Project Budget by Risk Level:")
print(df_clean.groupby('Risk_Level')['Project_Budget_USD'].mean())
```

```python
print("\nAverage Complexity Score by Project Type:")
print(df_clean.groupby('Project_Type')['Complexity_Score'].mean())
```

```python
print("\nAverage Budget per Team Member by Complexity Category:")
print(df_clean.groupby('Complexity_Category')['Budget_per_Team_Member'].mean())
```
